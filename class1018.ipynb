```python
def merge(A, left, mid, right):
    k = left
    i = left
    j = mid + 1
    while i <= mid and j <= right:
        if A[i] <= A[j]:
            sorted[k] = A[i]
            i, k = i + 1, k + 1
        else:
            sorted[k] = A[j]
            j, k = j + 1, k + 1
    
    if i > mid:
        sorted[k : k + right - k + 1] = A[j : right + 1]
    else:
        sorted[k : k + mid - i + 1] = A[i : mid + 1]

    A[left : right + 1] = sorted[left : right + 1]
```

```python
def merge_sort(A, left, right):
    if left < right:
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)
```

```python
def median_of_three(A, left, right):
    m = (left + right) // 2
    if ((A[left] < A[m] and A[m] < A[right]) or (A[right] < A[m] and A[m] < A[left])):
        return m
    if ((A[m] < A[left] and A[left] < A[right]) or (A[right] < A[left] and A[left] < A[m])):
        return left
    else:
        return right
```

```python
def partition (A, left, right):

    # 중간값을 피벗으로 선택
    pivot_index = median_of_three(A, left, right)
    # 피벗을 맨 왼쪽으로 이동
    A[left], A[pivot_index] = A[pivot_index], A[left]

    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high):
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high] > pivot : high -= 1

        if low < high:
            A[low], A[high] = A[high], A[low]
    
    if left != high:
        A[left], A[high] = A[high], A[left]
    return high
```

```python
def quick_sort(A, left, right):
    if left < right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid-1)
        quick_sort(A, mid + 1, right)
```

```python
data = [5, 3, 8, 4, 9, 1, 6, 2, 7] # 입력 리스트
print("Original :", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort :", data)
```
